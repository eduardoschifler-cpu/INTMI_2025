```python
#1. Crie uma lista com 15 números inteiros digitados pelo usuário. Em seguida, remova todos os números repetidos, mantendo a ordem original.
numeros = []
for _ in range(15):
    numeros.append(int(input("Digite um número: ")))

sem_repetidos = []
for n in numeros:
    if n not in sem_repetidos:
        sem_repetidos.append(n)

print("Lista sem repetidos:", sem_repetidos)
```

```python
#2. Escreva um programa que leia 10 nomes e 10 idades. Armazene os nomes em uma lista e as idades em outra. Em seguida, exiba o nome da pessoa mais velha.
nomes = []
idades = []

for _ in range(10):
    nomes.append(input("Nome: "))
    idades.append(int(input("Idade: ")))

mais_velho = max(idades)
indice = idades.index(mais_velho)

print("Pessoa mais velha:", nomes[indice])
```

```python
#3. Faça um programa que leia uma lista de 10 números e divida todos os valores ímpares por 3 e os pares por 2, substituindo os valores originais.
lista = []
for _ in range(10):
    lista.append(int(input("Número: ")))

for i in range(len(lista)):
    if lista[i] % 2 == 0:
        lista[i] /= 2
    else:
        lista[i] /= 3

print("Lista modificada:", lista)
```

```python
8
#5. Leia 10 números inteiros e armazene-os em uma lista. Em seguida, crie uma segunda lista apenas com os números primos da primeira.
def primo(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

numeros = [int(input("Número: ")) for _ in range(10)]
primos = [n for n in numeros if primo(n)]

print("Números primos:", primos)
```

```python
#11. Crie uma matriz 4x4 com valores fornecidos pelo usuário. Exiba a soma dos elementos da diagonal secundária.
matriz = []
for i in range(4):
    linha = []
    for j in range(4):
        linha.append(int(input(f"Elemento [{i}][{j}]: ")))
    matriz.append(linha)

soma = 0
for i in range(4):
    soma += matriz[i][3 - i]

print("Soma da diagonal secundária:", soma)
```

```python
7
#13. Solicite uma matriz 3x3 preenchida pelo usuário. Verifique se ela é uma matriz identidade.
matriz = []
for i in range(3):
    linha = []
    for j in range(3):
        linha.append(int(input(f"Elemento [{i}][{j}]: ")))
    matriz.append(linha)

identidade = True
for i in range(3):
    for j in range(3):
        if (i == j and matriz[i][j] != 1) or (i != j and matriz[i][j] != 0):
            identidade = False

print("É identidade?" , "Sim" if identidade else "Não")
```

```python
#14. Faça um programa que leia uma matriz 3x3 e verifique se ela é simétrica (igual à sua transposta).
matriz = []
for i in range(3):
    linha = []
    for j in range(3):
        linha.append(int(input(f"Elemento [{i}][{j}]: ")))
    matriz.append(linha)

simetrica = True
for i in range(3):
    for j in range(3):
        if matriz[i][j] != matriz[j][i]:
            simetrica = False

print("É simétrica?", "Sim" if simetrica else "Não")

```

```python
#16. Crie uma matriz 4x4 com números fornecidos pelo usuário. Encontre o menor valor de cada coluna e armazene em uma lista separada.
matriz = []
for i in range(4):
    linha = []
    for j in range(4):
        linha.append(float(input(f"Valor [{i},{j}]: ")))
    matriz.append(linha)

menores_colunas = []
for col in range(4):
    menores_colunas.append(min(matriz[i][col] for i in range(4)))

print("Menores valores de cada coluna:", menores_colunas)
```

```python
#17. Dada uma matriz 3x3 com valores reais, calcule o determinante (sem usar bibliotecas externas).
m = []
for i in range(3):
    m.append([float(x) for x in input().split()])

det = (m[0][0]*(m[1][1]*m[2][2] - m[1][2]*m[2][1])
     - m[0][1]*(m[1][0]*m[2][2] - m[1][2]*m[2][0])
     + m[0][2]*(m[1][0]*m[2][1] - m[1][1]*m[2][0]))

print("Determinante:", det)
```

```python
#18. Faça um programa que leia duas matrizes 3x3 e calcule a matriz soma e a matriz diferença.
A = []
B = []

print("Digite a matriz A:")
for _ in range(3):
    A.append([float(x) for x in input().split()])

print("Digite a matriz B:")
for _ in range(3):
    B.append([float(x) for x in input().split()])

soma = [[A[i][j] + B[i][j] for j in range(3)] for i in range(3)]
dif  = [[A[i][j] - B[i][j] for j in range(3)] for i in range(3)]

print("Matriz Soma:", soma)
print("Matriz Diferença:", dif)
```

```python
#25. Crie um dicionário que armazene o nome e o salário de 5 funcionários. Calcule o reajuste de 10% para salários inferiores a R$ 2.000 e atualize o dicionário.
func = {}

for i in range(5):
    nome = input(f"Nome {i+1}: ")
    salario = float(input("Salário: "))
    func[nome] = salario

for nome in func:
    if func[nome] < 2000:
        func[nome] *= 1.10

print(func)
```

```python
77#26. Faça um programa que armazene em um dicionário o nome de 5 alunos e uma lista com suas 3 notas. Depois, exiba o nome e a média de cada aluno.
alunos = {}

for i in range(5):
    nome = input(f"Aluno {i+1}: ")
    notas = [float(input("Nota: ")) for _ in range(3)]
    alunos[nome] = notas

for nome, notas in alunos.items():
    media = sum(notas) / 3
    print(nome, "-> Média:", round(media, 2))
```

```python
#29. Crie um dicionário onde as chaves são nomes de produtos e os valores são listas com preços praticados em três lojas diferentes. Encontre o produto com maior variação de preço.
produtos = {}

for i in range(3):
    nome = input("Produto: ")
    precos = [float(x) for x in input("3 preços: ").split()]
    produtos[nome] = precos

maior_var = None
produto_escolhido = ""

for nome, precos in produtos.items():
    variacao = max(precos) - min(precos)
    if maior_var is None or variacao > maior_var:
        maior_var = variacao
        produto_escolhido = nome

print("Produto com maior variação:", produto_escolhido)
```

```python
#30. Implemente um programa que leia uma frase e armazene em um dicionário o número de ocorrências de cada palavra (ignorando letras maiúsculas/minúsculas).
frase = input("Digite a frase: ").lower()

palavras = frase.split()

contagem = {}

for p in palavras:
    contagem[p] = contagem.get(p, 0) + 1

print(contagem)
```
